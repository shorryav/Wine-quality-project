#FIXED ACIDITY

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data, x = 'fixed acidity', color = 'red', edgecolor = 'linen', alpha = 0.5, bins = 5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.show()
```

```python
round(data['fixed acidity'].mean(),2)
```

```python
data['fixed acidity'].median()
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data, x = 'fixed acidity', color = 'red', edgecolor = 'linen', alpha = 0.5, bins = 5)
plt.title("Hostogram Of Fixed Acidity")
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.vlines(data['fixed acidity'].mean(), color = 'blue', ymin = 0, ymax = 4000, label = 'mean')
plt.vlines(data['fixed acidity'].median(), color = 'green', ymin = 0, ymax = 4000, label = 'median')
plt.legend()
plt.show()
```

# OBSERVATION


1. MEAN AND MEDIAN IS VERY CLOSE TO EACH OTHER
2. IT IS EQUALLY DISTRIBUTED
3. WE CAN CHOOSE EITHER MEAN OR MEDIAN AS A CENTRAL TENDENCY
4. IT IS CLEAR REPRESENTAION OF DATA OF MEAN AND MEDIAN



# VOLATILE ACIDITY

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data, x = 'volatile acidity', color = 'red', edgecolor = 'linen', alpha = 0.5, bins = 5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.show()
```

```python
round(data['volatile acidity'].mean(),2)
```

```python
data['volatile acidity'].median()
```

```python
plt.figure(figsize = (11,6))
sns.histplot(data = data, x = 'volatile acidity', color = 'blue', edgecolor = 'linen', alpha = 0.5, bins = 5)
plt.title("Hostogram Of Volatile Acidity")
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.vlines(data['volatile acidity'].mean(), color = 'red', ymin = 0, ymax = 4000, label = 'mean')
plt.vlines(data['volatile acidity'].median(), color = 'green', ymin = 0, ymax = 4000, label = 'median')
plt.legend()
plt.show()
```

# OBSERVATION


1.   WE HAVE OBSERVED THAT THE GRAPH IS NOT EQUALLY DISTRIBUTED AND IT IS SKEWED A LITTLE TOWARDS THE RIGHT.
2. WE HAVE OBSERVED THAT MEAN AND MEDIAN IS CLOSE TO EACH OTHER AND THE DIFFERENCE BETWEEN THEM IS VERY SMALL.
3. AS WE HVE OBSERVED THE SKEWNESS, SO WE CAN CHECK THE DISTRIBUTION BY USING THE DISTRIBUTION PLOT.


```python
plt.figure(figsize = (11,6))
sns.distplot(data['volatile acidity'], color = 'blue')
plt.title("Distribution of Volatile Acidity")
plt.xlabel('volatile acidity')
plt.ylabel('Density')
plt.show()
```

```python
round(data['volatile acidity'].skew(),2)
```

# OBSERVATION
WE HAVE CHECK THAT THE ABOVE GRAPH SHPWS THE NORMAL DISTRIBUTION

# RESIDUAL SUGAR

```python
plt.figure(figsize =(11,6))
sns.histplot(data = data, x = 'residual sugar', alpha = 0.5, bins = 5, color = 'pink', edgecolor = 'linen')
plt.title("HISTOGRAM OF RESIDUAL SUGAR")
plt.xlabel('residual sugar')
plt.ylabel('count')
plt.show()
```

```python
round(data['residual sugar'].mean(),2)
```

```python
data['residual sugar'].median()
```

```python
plt.figure(figsize =(11,6))
sns.histplot(data = data, x = 'residual sugar', alpha = 0.5, bins = 5, color = 'pink', edgecolor = 'linen')
plt.title("HISTOGRAM OF RESIDUAL SUGAR")
plt.xlabel('residual sugar')
plt.ylabel('count')
plt.vlines(data['residual sugar'].mean(), color = 'red', ymin = 0, ymax = 4000, label = 'mean')
plt.vlines(data['residual sugar'].median(), color = 'blue', ymin = 0, ymax = 4000, label = 'median')
plt.legend()
plt.show()
```

* mean and median is close to each other and the difference between them is very small



```python
plt.figure(figsize = (11,6))
sns.distplot(data['residual sugar'], color = 'blue')
plt.title("DISTRIBUTION OF RESIDUAL PLOT")
plt.xlabel('quality')
plt.ylabel('Density')
plt.show()
```
